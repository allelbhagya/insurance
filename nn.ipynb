```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./insurance.csv')
data.head()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'].values.reshape((-1,1)))
data['smoker'] = le.fit_transform(data['smoker'].values.reshape((-1,1)))
data['region'] = le.fit_transform(data['region'].values.reshape((-1,1)))
data
```

```python
data.dtypes
```

```python
X = data.drop(['charges'],axis = 1)
y = data['charges']
X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
X_scaled = std.fit_transform(X)
y_scaled = std.fit_transform(y.values.reshape((-1,1)))
X_scaled.shape, y_scaled.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
X_train.shape, y_train.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(12, input_shape = (X_train.shape[1],), activation = 'relu' ))
model.add(Dense(32, activation = 'sigmoid' ))
model.add(Dense(16, activation = 'relu' ))
model.add(Dense(4, activation = 'sigmoid' ))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'sgd', metrics = ['mae', 'mse'])
model.summary()
```

```python
model.fit(X_train,
          y_train,
          epochs = 200,
          verbose = 1,
          validation_split = 0.2)
```

```python
from sklearn.metrics import r2_score

pred = model.predict(X_test)
r2_score(y_test, pred)
```
